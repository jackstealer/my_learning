```python



import cv2
import numpy as np
```

```python
haar_features=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img=cv2.imread('new.jpg')
```

```python
# Corrected code with proper method name
haar_features.detectMultiScale(img)  # Note the capital 'S' in 'Scale'
```

```python
faces=haar_features.detectMultiScale(img)
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,225,0),2)
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
