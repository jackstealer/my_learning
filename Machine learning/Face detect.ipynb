```python
#Face detection
import cv2
import numpy as np
```

```python
haar_features = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('face.jpeg')
```

```python
haar_features.detectMultiScale(img)
faces=haar_features.detectMultiScale(img)
for (x,y,w,h) in faces:
    # img- Original image on which we want to draw rectangle
    # (x,y) - ( face top-left section) co-ordiante where face it detected
    #  (x+w, y+h) - (bottom-right corner)
    #  (0,255,0) - green color
    #  2- thickness of border
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
