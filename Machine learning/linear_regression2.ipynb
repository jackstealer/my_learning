```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5, 6], dtype=float)
y = np.array([3, 4, 6, 8, 10, 12], dtype=float)
```

```python
x = (x - x.mean()) / x.std()
y = (y - y.mean()) / y.std()

n = len(x)
```

```python
def gradient_descent(x, y, epochs, alpha):
    slope = 0
    intercept = 0
    
    for i in range(epochs):
        y_pred = slope * x + intercept
        
        error = y_pred - y
        
        # Gradients
        d_slope = (2/n) * np.dot(error, x)
        d_intercept = (2/n) * np.sum(error)
        
        # Update
        slope = slope - alpha * d_slope
        intercept = intercept - alpha * d_intercept
        
    return slope, intercept
```

```python
epochs = 1000
alpha = 0.01

slope, intercept = gradient_descent(x, y, epochs, alpha)

print("Slope:", slope)
print("Intercept:", intercept)
```

```python
y_pred = slope * x + intercept
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
```

```python
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("Experience (Normalized)")
plt.ylabel("Salary (Normalized)")
plt.show()
```
