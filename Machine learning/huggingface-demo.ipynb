```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation', model='gpt2')
```

```python
result = generator("that person is in the", max_length=50, num_return_sequences=1)
print(result[0]['generated_text'])
```

```python
classifier = pipeline('sentiment-analysis',model ='gpt2')
result = classifier("I always love using Hugging Face!")
print(result) 
```
