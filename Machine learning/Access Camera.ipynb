```python
import cv2
```

```python
cap=cv2.VideoCapture(0)
while True:
    #get frames from video
    ret, frame=cap.read()
    #break if video ends or fails
    if not ret:
        break
    cv2.imshow("My camera",frame)
    key=cv2.waitKey(25) & 0xFF  #milliseconds
    if key==27: #27: ESC
        break 
cap.release()
cv2.destroyAllWindows()
```
