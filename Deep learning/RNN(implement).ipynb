```python
tokenizer =Tokenizer()
tokenizer.fit_on_texts(text)
```

```python
sequences = tokenizer.texts_to_sequences(text)
print(sequences)
```

```python
X = []
y =[]
for seq in sequences:
    for i in range(1, len(seq)):
        X.append(seq[:i])
        y.append(seq[i])
```

```python
X
```

```python
X = pad_sequences(X)
```

```python
X
```

```python
y = np.array(y)
```

```python
y
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
model = Sequential([
    Embedding(vocab_size, 15, input_length = X.shape[1]),
    SimpleRNN(32),
    Dense(vocab_size, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
)
```

```python
model.fit(X,y,
         epochs=1000)
```

```python
input_text = "i want"
seq = tokenizer.texts_to_sequences([input_text])[0]
seq = pad_sequences([seq], maxlen =X.shape[1])
pred = model.predict(seq)
```

```python
pred
```

```python
predicted_index=np.argmax(pred)
for word, index in tokenizer.word_index.items():
    if index == predicted_index:
        print("Next word:", word)
```

```python
def predict_next_word(model,tokenizer,input_text,n_words):
  seq = tokenizer.texts_to_sequences([input_text])[0]
  seq = pad_sequences([seq],maxlen=X.shape[1])
  pred=model.predict(seq)
  predicted_index = np.argmax(pred)
  for word,index in tokenizer.word_index.items():
    if index == predicted_index:
      return word
```

```python
text = "i want"
for _ in range(5):
  next_word = predict_next_word(model,tokenizer,text,X.shape[1])
  text +=" "+next_word

print(text)
```
