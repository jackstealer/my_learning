```python
import numpy as np


X = np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]])
y = np.array([[1],[1],[0]])

def sigmoid_activation(x):
  return 1 / (1 + np.exp(-x))

def derivative_sigmoid(x):
    return x * (1 - x)
```

```python
input_neurons = X.shape[1]
hidden_neuron = 3
output_neuron = 1
epochs = 100000
learning_rate = 0.1
```

```python
# weights and bies initialization
wh = np.random.uniform(size=(input_neurons, hidden_neuron))
bh = np.random.uniform(size=(1, hidden_neuron))
wout = np.random.uniform(size=(hidden_neuron, output_neuron))
bout = np.random.uniform(size=(1, output_neuron))
```

```python

for i in range(epochs):
  hidden_layer_input = np.dot(X, wh) + bh
  hidden_layer_activation = sigmoid_activation(hidden_layer_input)
  output_layer = np.dot(hidden_layer_activation, wout) + bout
  output = sigmoid_activation(output_layer)

  output_error = y - output
  output_slope = derivative(output)
  delta_output = output_error * output_slope

  hidden_loss = delta_output.dot(wout.T)
  hidden_slope = derivative(hidden_layer_activation)
  hidden_delta = hidden_slope * hidden_loss

  wout = wout + hidden_layer_activation.T.dot(delta_output) * learning_rate
  wh = wh + X.T.dot(hidden_delta) * learning_rate
  bout = bout + np.sum(delta_output, axis=0, keepdims=True) * learning_rate
  bh = bh + np.sum(delta_output, axis=0, keepdims=True) * learning_rate

output
```
