```python
import nltk
#4 things to do with nltk
# a. tokenization
# b. stopword removal
# c. stemming
# d. lemmatization
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download('punkt_tab')
nltk.download("wordnet")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
raw_text ="Hey! I am learning the python programming, Python has slow execution than java and cpp...!!"
```

```python
print(raw_text.split())
```

```python
#step -1 : lowercase
raw_text = raw_text.lower()
print(raw_text)
```

```python
#step 2 : tokenizaton
tokens = word_tokenize(raw_text)
```

```python
tokens
```

```python
# step-3 : remove stopwords
english_stopwords =stopwords.words("english")
```

```python
filtered_tokens =[]
for word in tokens:
    if word not in english_stopwords:
        filtered_tokens.append(word)

print(filtered_tokens)
```

```python
# remove punctuations
import string
punctuations = string.punctuation
print(punctuations)
```

```python
clean_tokens = [word for word in filtered_tokens if word not in punctuations]
print(clean_tokens)
```

```python
stem = PorterStemmer()
# stem.stem("playing")
# stem.stem("wathing")
# stem.stem("flying")
stem.stem("bought")
```

```python
wnet = WordNetLemmatizer()
wnet.lemmatize("playing", "v")
wnet.lemmatize("removing", "v")
```

```python
lemmatized_words =[]
for word in clean_tokens:
    lemmatized_words.append(wnet.lemmatize(word,"v"))

print(lemmatized_words)
```

```python
final_tokens =[]
for word in lemmatized_words:
    if word.isalpha():
        final_tokens.append(word)

print(final_tokens)
```

```python
cleaned_text =" ".join(final_tokens)
print(cleaned_text)
```
