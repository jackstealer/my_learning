```python
# hii
# hellow
# how are you
# what are you doing man
# what is the origin of the universe
# did you really realize that how the human species survived
text ="Hey ! I am learning the python programming languages..!!"
print("Original text :",text)
```

```python
#step-1 conver to lowercase
text=text.lower()
print(text)
```

```python
#step-2 remove punctuation marks
#re-regex-regular expression
import re
text=re.sub(r'[^\w\s]', '',text) #remove everthinng except letters and spaces
print(text)
```

```python
#step-3 tokenization
tokens =text.split()
print(tokens)
```

```python
#remove stopwords
stopwords =["a", "am" , "is", "are", "i", "the","and"]
filtered_tokens=[]
for word in tokens:
    if word not in stopwords:
        filtered_tokens.append(word)

print("After removeing stopwords: ",filtered_tokens)
```

```python
def stem(word):
    if word.endswith("ing"):
        return word[:-3]
    return word
stemmed_words =[]
for word in filtered_tokens:
    stemmed_words.append(stem(word))
print("after stemming: " , stemmed_words)
```

```python
#NLP PHASE 2
#will convert text ->count
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
sentences=[
    "i am learning python",
    "we are happy today",
    "i am sad today",
    "i am learning machine learning"
]
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(sentences) 
print(vectorizer.get_feature_names_out())
#each row is a sentence
#each column is a word
#each value is a count
```

```python
print(X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences=[
    "i am learning python",
    "we are happy today",
    "i am sad today",
    "i am learning machine learning"
]
```

```python
vectorizer =TfidfVectorizer()
X= vectorizer.fit_transform(sentences)
#buid vacabulary
#calculate TF
#calculate idf - impact of each word
#multiples - tf x idf
#produces matrix
```

```python
print(X.toarray())
```

```python
print(vectorizer.get_feature_names_out())
```
